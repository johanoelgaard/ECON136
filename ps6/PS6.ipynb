```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in excel file
df_org = pd.read_excel('returns.xls', header=2)
df = df_org.iloc[:,8:].dropna()
df.tail()
```

```python
# plot Weight in Growth as a function of Std.Dev and mark points with a circle
plt.figure(figsize=(10,6))
plt.plot(df['Weight in Growth'],df['Std.Dev'], 'o-')
plt.ylabel('Standard Deviation', fontsize=16)
plt.xlabel('Weight in Growth', fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)
plt.grid(True)
plt.tight_layout()
plt.savefig('weight_stddev.png')
plt.show()
```

```python
# plot Weight in Growth as a function of Std.Dev and mark points with a circle
plt.figure(figsize=(10,6))
plt.plot(df['Std.Dev'], df['Exp Return'], marker='o', label='Minimum Variance Frotier')
plt.plot(df['Std.Dev'][:df['Std.Dev'].idxmin()+1], df['Exp Return'][:df['Std.Dev'].idxmin()+1], color='r', label='Efficient Frontier')
plt.plot(df['Std.Dev'].min(), df['Exp Return'][df['Std.Dev'].idxmin()], 'o', color='orange', label=f'Minimum Variance Portfolio,\n $w_G=${df["Weight in Growth"][df["Std.Dev"].idxmin()]:.2f}, $\sigma_p=${df["Std.Dev"].min():.4f}, $\mu_p=${df["Exp Return"][df["Std.Dev"].idxmin()]:.4f}')
plt.legend(fontsize=16)
plt.xlabel('Standard Deviation', fontsize=16)
plt.ylabel('Expected Return', fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)
plt.grid(True)
plt.tight_layout()
plt.savefig('return_stddev.png')
plt.show()
```
