```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in excel file
df = pd.read_excel('returns.xls', header=0)

df_1dollar = pd.concat((df.iloc[:,0],df.iloc[:,19:24]), axis=1)

# promote first row to column names
df_1dollar.columns = df_1dollar.iloc[0]
df_1dollar = df_1dollar[:-3]
df_1dollar = df_1dollar.drop(0).set_index('Year')
df_1dollar
```

```python
# plot the value 1 dollar invested in 1927
for col in df_1dollar.columns:
    df_1dollar[col].plot(legend=True, grid=True, figsize=(10,6))
plt.yscale('log')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Nominal value of $1 dollar', fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)
plt.legend(fontsize=16)
plt.tight_layout()
plt.savefig('1dollar_invested_1927.png')
plt.show()
```
